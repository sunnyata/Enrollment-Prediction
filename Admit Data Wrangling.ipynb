```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
current_directory = os.getcwd()
print(current_directory) 
pd.set_option("display.max_rows", 999)
pd.set_option("display.max_columns", 999)
pd.options.display.max_seq_items = 2000
```

```python
# Reading in all app data and parsing date columns as datetime
date_cols = ['student_complete_date','application_date', 'app_complete_date','admit_date','enrollment_deposit_dt','namu_award_date']
s1 = pd.read_csv("App-UGRD_Master-Data_INTL_Fall2021_Anonymized.csv", encoding='latin-1', parse_dates=date_cols)
s2 = pd.read_csv("App-UGRD_Master-Data_INTL_Fall2022_Anonymized.csv", encoding='latin-1', parse_dates=date_cols)
s3 = pd.read_csv("App-UGRD_Master-Data_INTL_Fall2023_Anonymized.csv", encoding='latin-1', parse_dates=date_cols)
s4 = pd.read_csv("App-UGRD_Master-Data_INTL_Fall2024_Anonymized.csv", encoding='latin-1', parse_dates=date_cols)
```

```python
# Filtering each year's app data to exclude apps started later than point-in-cycle (June 6)
s1 = s1[(s1['application_date']<pd.Timestamp(2021,6,7))]
s1['application_date'].max()
```

```python
s3 = s3[(s3['application_date']<pd.Timestamp(2023,6,7))]
s3['application_date'].max()
```

```python
# Concatenating all app data 2021-4
frames = [s1, s2, s3, s4]
apps = pd.concat(frames)
apps.drop_duplicates(inplace=True)
apps.to_csv("AppsALL.csv", index=False)
apps[apps.duplicated(subset="app_id")]
```

```python
date_cols = ['contact_createddate', 'lead_createddate_FLS','othersource_created_dt', 'case_created_dt','event_start_date','application_date','first_date']
sources = pd.read_csv("TempAll.csv", encoding='latin-1', parse_dates=date_cols)
sources.sort_values(by='application_date', ascending=True, inplace=True)
sources.drop_duplicates(subset=['app_id'], keep='last', inplace= True)
```

```python
# Merging first source data with app data
temps = pd.merge(apps, sources[['app_id','contact_createddate','first_date', 'first_source_type', 'first_source','first_source_group', 'cold_lead']], on='app_id', how='left')
```

```python
temps.shape
```

```python
# Adding first date and source=stealth where it was missing from sources/lead data
mask = temps['first_date'].isnull()
temps['first_date'] =  np.where(mask, temps['application_date'], temps['first_date'])
temps['first_source_type'] =  np.where(mask, "Stealth App", temps['first_source_type'])
temps['first_source'] =  np.where(mask, "Stealth App", temps['first_source'])
temps['cold_lead'] =  np.where(mask, 0, temps['cold_lead'])
```

```python
# Cleaning file by renaming and removing clumns to prepare for R
temps.rename(columns={'first_date': 'first_source_date','app_complete_count': 'app_complete','admit_count':'admitted', 'enrl_count':'enrolled'}, inplace=True)
temps.drop(temps.iloc[:, 2:7], inplace=True, axis=1)
temps.drop(['app_count', 'app_activity_count', 'student_complete_date','conditional_admit_count','deny_count','enrollment_deposit_status'],inplace=True, axis=1)
temps.drop(['housing_confirmed_count','residency_status','market','ethnicity','hs_name','hs_code','hs_city','postsec_city','postsec_country','ci_band_domestic'],inplace=True, axis=1)
temps.drop(['namu_transfer_amount',	'namu_transfer_offer_dt','namu_transfer_scholarship','move_to_online','visa_permit','visa_descr'],inplace=True, axis=1)
temps.drop(['honors_app_date','honors_denied','honors_enrolled','ci_score_intl','namu_scholarship','intl_remote','enrollment_deposit_dt'],inplace=True, axis=1)
```

```python
mask = temps['ugrd_intl_gpa'].isnull()

temps['GPA'] = (temps[['hs_gpa', 'trn_gpa', 'abor_gpa', 'self_gpa', 'cbov_gpa']]
             .apply(pd.to_numeric)
             .max(axis=1))   # <-- difference here
```

```python
# temps.to_csv("AppsR.csv", index=False)
```

```python
temps['ugrd_intl_gpa'] =  np.where(mask, temps['GPA'], temps['ugrd_intl_gpa'])
# temps.to_csv("AppsR.csv", index=False)
```

```python
temps.drop(['hs_gpa','abor_gpa','self_gpa','cbov_gpa','trn_gpa','trn_gpa_band','GPA'],inplace=True, axis=1)
temps = temps[temps['application_status'] != "Application Received - Incomplete"]
# temps.to_csv("AppsR.csv", index=False)
```

```python
temps.drop(['namu_award_date','honors_admit', 'kaplan_university_scholarship','kaplan_international_award','contact_createddate'],inplace=True, axis=1)
temps.rename(columns={'hs_country':'us_hs_flag', 'postsec_inst_name': 'postsec_flag'},inplace=True)

# Creating new binary and categorical predictors
temps['us_hs_flag'] =  np.where(temps['us_hs_flag'] == "USA", 1, 0)
temps['honors_applied'] =  np.where(temps['honors_applied'].isnull(), 0, temps['honors_applied'])
temps['postsec_flag'] =  np.where(temps['postsec_flag'].notnull(), 1, 0)
temps['kaplan_flag'] =  np.where(temps['kaplan_flag'].notnull(), 1, 0)
temps['defer_flag'] = np.where(temps['defer_flag'].notnull(), 1, 0)
temps['common_app'] =  np.where(temps['common_app'].isnull(), 0, temps['common_app'])
temps['namu_amount'] =  np.where(temps['namu_amount'].notnull(), temps['namu_amount'], 0)
temps['first_source_month'] = temps['first_source_date'].dt.month
temps['app_month'] = temps['application_date'].dt.month
temps['app_comp_month'] = temps['app_complete_date'].dt.month
temps['admit_month'] = temps['admit_date'].dt.month
# Imputing some missing values (12 total rows)
temps['app_comp_month'] = np.where(temps['app_comp_month'].notnull(), temps['app_comp_month'], 0)
# Imputing some missing values (3050 total rows)
temps['admit_month']  = np.where(temps['admit_month'].notnull(), temps['admit_month'], 0)
# Fixing College nomenclature for WPC
temps['college'] = np.where(temps['college'] == "W. P. Carey School of Business", "W.P. Carey School of Business", temps['college'])
```

```python
# Choosing 9/7 to avoid creating negative values for same apps completed very late in cycle
temps['app_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2021", (pd.Timestamp(2021,9,8) - temps['application_date']).dt.days, 0)
temps['app_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2022", (pd.Timestamp(2022,9,8) - temps['application_date']).dt.days, temps['app_date_diff'])
temps['app_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2023", (pd.Timestamp(2023,9,8) - temps['application_date']).dt.days, temps['app_date_diff'])
temps['app_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2024", (pd.Timestamp(2024,9,8) - temps['application_date']).dt.days, temps['app_date_diff'])

temps['appcomp_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2021", (pd.Timestamp(2021,9,8) - temps['app_complete_date']).dt.days, 0)
temps['appcomp_date_diff']=  np.where(temps['Term_Group'] == "Fall 2022", (pd.Timestamp(2022,9,8) - temps['app_complete_date']).dt.days, temps['appcomp_date_diff'])
temps['appcomp_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2023", (pd.Timestamp(2023,9,8) - temps['app_complete_date']).dt.days, temps['appcomp_date_diff'])
temps['appcomp_date_diff'] =  np.where(temps['Term_Group'] == "Fall 2024", (pd.Timestamp(2024,9,8) - temps['app_complete_date']).dt.days, temps['appcomp_date_diff'])

# temps['application_date'] = (df['date2'] - df['date1']).dt.days
```

```python
# Dropping transformed date columns and EPR score columns which are rather sparse to use for prediction -- 14,611 than have no EPR score at all
temps.drop(['application_status','ielts_score','toefl_comp','det_score', 'admit_date', 'app_complete_date', 'application_date', 'first_source_date','first_source_group'],inplace=True, axis=1)
# temps.to_csv("AppsR.csv", index=False)
```

```python
# looking at NA values and imputing data
temps.isna().sum()
```

```python
# Using the average difference between when the app was submitted and when it was completed to fill in 12 missing values for appcomp_date_diff
m = round((temps.app_date_diff- temps.appcomp_date_diff).mean())
temps['appcomp_date_diff']  = np.where(temps['appcomp_date_diff'].isnull(), m, temps['appcomp_date_diff'])

# Adding two more feautres
temps['app_appcomp_diff'] = temps['app_date_diff'] - temps['appcomp_date_diff']
```

```python
temps.isna().sum()
```

```python
m = round(temps.age.mean())
temps['age']  = np.where(temps['age'] == 0, m, temps['age'])
temps['age']  = np.where(temps['age'].isnull(), m, temps['age'])
```

```python
m = round(temps.ugrd_intl_gpa.mean(), 2)
temps['ugrd_intl_gpa']  = np.where(temps['ugrd_intl_gpa'].isnull(), m, temps['ugrd_intl_gpa'])
temps['home_city']  = temps['home_city'].str.replace(r"'", '', regex=False)
temps['citzn_country']  = temps['citzn_country'].str.replace(r"'", '', regex=False)
temps['home_country']  = np.where(temps['home_country'] == '-', "Unknown", temps['home_country'])
temps['citzn_country']  = np.where(temps['citzn_country'] == '-', "Unknown", temps['citzn_country'])
temps['home_city']  = np.where(temps['home_city'] == '-', "Unknown", temps['home_city'])
temps['home_city'] = np.where(temps['app_id'] == 'ANXKsaOaYyMree094huvrg==', "CAIRO", temps['home_city'])
temps['intl_financial_guarantee_flg'] = np.where(temps['intl_financial_guarantee_flg'] == 'Y', 1, 0)
temps['intl_sponsored_flg'] = np.where(temps['intl_sponsored_flg'] == 'Y', 1, 0)
temps['first_gen_flg'] = np.where(temps['first_gen_flg'] == 'Y', 1, 0)
temps['college'] =  np.where(temps['college'] == 'College of Design', "Herberger Institute", temps['college'])
temps['major'] =  np.where(temps['major'] == 'Performing Arts', "Art/Design", temps['major'])

temps['age'] = np.where(temps['age'] > 18, 1, 0)
mask = temps['major'].isin(['Liberal Studies', 'Media','Crime/Law', 'Education','History','English','Language','Non-Degree Seeking', 'Communication'])
temps['major'] =  np.where(mask, "Other", temps['major'])

temps['campus'] =  np.where(temps['campus'] == 'LOSAN', "OTHER", temps['campus'])

temps = temps[temps['gender'].isin(['M','F'])]

temps = temps[temps['campus'].isin(['TEMPE','POLY','WEST','DTPHX','CALHC'])]

temps = temps[temps['student_type'].isin(['FTF','TRN'])]
```

```python
# Country and City are variables are categorical with too high cardinality!  
# To keep the dimensionality of the dataset down in one-hot-encoding later, I opt to frequency encode them
temps.rename(columns={'age':'above_18'},inplace=True)

citzn_freq = temps['citzn_country'].value_counts(normalize=True)
temps['citzn_country_freq'] = temps['citzn_country'].map(citzn_freq)

home_city_freq = temps['home_city'].value_counts(normalize=True)
temps['home_city_freq'] = temps['home_city'].map(home_city_freq)

home_country_freq = temps['home_country'].value_counts(normalize=True)
temps['home_country_freq'] = temps['home_country'].map(home_country_freq)

temps.drop(['citzn_country','home_city','home_country'],inplace=True, axis=1)
temps.to_csv("AppsR.csv", index=False)
```
